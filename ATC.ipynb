```python
"""
Automatic Ticket Classification (Many-to-One LSTM) + LLM reply (template)
Run in Colab / Jupyter. Install missing packages first:
!pip install datasets tensorflow scikit-learn matplotlib
"""
#Embedding (128)

#Bidirectional LSTM (128)

#Dense (64)

#Softmax
# -----------------------
# STEP 0: Imports & setup
# -----------------------
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from datasets import load_dataset
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import json
import requests   # used by the example Gemini function (replace if you have official SDK)

SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# -----------------------
# STEP 1: Load dataset
# -----------------------
# Template suggests using HuggingFace dataset: Tobi-Bueck/customer-support-tickets.
ds = load_dataset("Tobi-Bueck/customer-support-tickets")

# dataset has train/test split depending on dataset; inspect
print(ds)

# We'll use the 'train' split if present, otherwise combine.
if "train" in ds:
    data = ds["train"]
else:
    # fallback to first split
    first_split = list(ds.keys())[0]
    data = ds[first_split]

# Quick preview
print("Example row:", data[0])

# Keep only necessary fields and remove examples without body or queue
texts = []
labels = []
for row in data:
    body = row.get("body") or ""
    queue = row.get("queue") or None
    if body and queue:
        texts.append(body)
        labels.append(queue)

print("Loaded texts:", len(texts))

# -----------------------
# STEP 2: Preprocessing
# -----------------------
# Basic cleaning function (customize to your needs)
# =============================================================
# FULL CLEANING FUNCTION FOR CUSTOMER SUPPORT TICKETS
# =============================================================
import re
import html
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_ticket(text: str) -> str:
    """Clean customer support ticket text with normalization, stopwords removal, and lemmatization"""
    if not isinstance(text, str):
        return ""
    
    # Basic text cleaning
    text = html.unescape(text)
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text)  # remove URLs
    text = re.sub(r"\S+@\S+", " ", text)                 # remove emails
    text = re.sub(r"\d+", " <num> ", text)              # replace numbers
    text = re.sub(r"[^a-zA-Z0-9\s\.\,\?\!\-_/]", " ", text)  # remove unwanted characters
    text = re.sub(r"\s+", " ", text).strip()            # remove extra spaces
    
    # Normalize common abbreviations
    text = text.replace("pls", "please")
    text = text.replace("asap", "as soon as possible")
    text = text.replace("cant", "cannot")
    text = text.replace("n't", " not")
    text = text.replace("u ", "you ")
    text = text.replace(" ur ", " your ")
    
    # Tokenization + stopword removal + lemmatization
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    text = " ".join(words)
    
    return text

# =============================================================
# APPLY CLEANING TO DATASET
# =============================================================
texts, labels = [], []

for row in data:  # assuming `data` is a list of dicts with keys 'body' and 'queue'
    body = row.get("body")
    queue = row.get("queue")
    if isinstance(body, str) and isinstance(queue, str) and body.strip() and queue.strip():
        texts.append(clean_ticket(body.strip()))
        labels.append(queue.strip())

print(f"✅ Cleaned dataset size: {len(texts)} texts, {len(labels)} labels")

# Label encoding
le = LabelEncoder()
y = le.fit_transform(labels)
num_classes = len(le.classes_)
print("Classes:", le.classes_)

# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(texts, y, test_size=0.15, stratify=y, random_state=SEED)

# Tokenization
vocab_size = 50000   # tune as needed
oov_tok = "<OOV>"
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(X_train)

# Sequences & padding
max_len = 200   # tune; can use median/percentile of lengths
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post', truncating='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=max_len, padding='post', truncating='post')

print("Shapes:", X_train_pad.shape, X_val_pad.shape)

# Convert labels to categorical if using categorical_crossentropy
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_val_cat = tf.keras.utils.to_categorical(y_val, num_classes=num_classes)

# -----------------------
# STEP 3: Build the Many-to-One RNN (LSTM)
# -----------------------
embedding_dim = 128

def build_model(vocab_size, embedding_dim, max_len, num_classes, recurrent_units=128, dropout_rate=0.5):
    model = Sequential()
    # Include input_length here:
    model.add(Embedding(input_dim=vocab_size,
                        output_dim=embedding_dim,
                        input_length=max_len))
    model.add(Bidirectional(LSTM(recurrent_units, return_sequences=False)))
    model.add(Dropout(dropout_rate))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))
    return model

model = build_model(vocab_size, embedding_dim, max_len, num_classes)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Option 1: Run a dummy input once to build it
model(tf.zeros((1, max_len)))

# Option 2: Or call model.build((None, max_len))
# model.build((None, max_len))

model.summary()

# -----------------------
# STEP 4: Callbacks
# -----------------------
os.makedirs("models", exist_ok=True)
checkpoint_path = "models/best_lstm.h5"
callbacks = [
    ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
]

# -----------------------
# STEP 5: Train
# -----------------------
batch_size = 64
epochs = 10

history = model.fit(
    X_train_pad, y_train_cat,
    validation_data=(X_val_pad, y_val_cat),
    epochs=epochs,
    batch_size=batch_size,
    callbacks=callbacks
)

# -----------------------
# STEP 6: Evaluation
# -----------------------
# Load best model weights (ModelCheckpoint already saved best)
if os.path.exists(checkpoint_path):
    model.load_weights(checkpoint_path)

# Predict on validation set
y_pred_prob = model.predict(X_val_pad)
y_pred = np.argmax(y_pred_prob, axis=1)

print("Classification report:")
print(classification_report(y_val, y_pred, target_names=le.classes_, zero_division=0))

# Confusion matrix (can visualize)
cm = confusion_matrix(y_val, y_pred)
print("Confusion matrix shape:", cm.shape)

# Plot training curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend(); plt.title("Loss")
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend(); plt.title("Accuracy")
plt.show()
```

```python
# =======================================
# CUSTOMER SUPPORT TICKET CLASSIFIER (v2)
# =======================================
#Embedding size increased (256 → richer word vectors)
#Max sequence length increased to 300 tokens
#Deeper Dense layers (256 → 128)
#High dropout (0.4)
#Learning rate improved
#25 epochs
#Stronger callbacks

# ✅ Imports
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from datasets import load_dataset
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import re
import html
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# ✅ Setup
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# =======================================
# STEP 1: Load dataset
# =======================================
ds = load_dataset("Tobi-Bueck/customer-support-tickets")
data = ds["train"] if "train" in ds else ds[list(ds.keys())[0]]

texts, labels = [], []
for row in data:
    body = row.get("body") or ""
    queue = row.get("queue") or None
    if body and queue:
        texts.append(body)
        labels.append(queue)

print(f"Loaded {len(texts)} samples")

# =======================================
# STEP 2: Preprocessing
# =======================================
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_ticket(text: str) -> str:
    """Clean and normalize ticket text."""
    if not isinstance(text, str):
        return ""
    text = html.unescape(text)
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text)
    text = re.sub(r"\S+@\S+", " ", text)
    text = re.sub(r"\d+", " <num> ", text)
    text = re.sub(r"[^a-zA-Z0-9\s\.\,\?\!\-_/]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    # Abbreviation normalization
    text = text.replace("pls", "please").replace("asap", "as soon as possible")
    text = text.replace("cant", "cannot").replace("n't", " not")
    text = text.replace("u ", "you ").replace(" ur ", " your ")
    # Lemmatize and remove stopwords
    words = [lemmatizer.lemmatize(w) for w in text.split() if w not in stop_words]
    return " ".join(words)

texts = [clean_ticket(t) for t in texts]

# =======================================
# STEP 3: Label encoding
# =======================================
le = LabelEncoder()
y = le.fit_transform(labels)
num_classes = len(le.classes_)
print("Classes:", num_classes, le.classes_[:10], "...")

# =======================================
# STEP 4: Tokenization and Padding
# =======================================
X_train, X_val, y_train, y_val = train_test_split(
    texts, y, test_size=0.15, stratify=y, random_state=SEED
)

vocab_size = 60000
oov_tok = "<OOV>"
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(X_train)

max_len = 300  # longer sequences capture more context
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post', truncating='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=max_len, padding='post', truncating='post')

y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_val_cat = tf.keras.utils.to_categorical(y_val, num_classes=num_classes)

# =======================================
# STEP 5: Handle class imbalance
# =======================================
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = dict(enumerate(class_weights))

# =======================================
# STEP 6: Build Improved BiLSTM Model
# =======================================
#Embedding size increased (256 → richer word vectors)
#Max sequence length increased to 300 tokens
#Deeper Dense layers (256 → 128)
#High dropout (0.4)
#Learning rate improved
#25 epochs
#Stronger callbacks
def build_bilstm(vocab_size, embedding_dim, max_len, num_classes):
    model = Sequential()
    model.add(Embedding(vocab_size, embedding_dim, input_length=max_len, trainable=True))
    model.add(Bidirectional(LSTM(256, return_sequences=True)))
    model.add(Dropout(0.4))
    model.add(Bidirectional(LSTM(128)))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    return model

embedding_dim = 256
model = build_bilstm(vocab_size, embedding_dim, max_len, num_classes)

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

model(tf.zeros((1, max_len)))  # build model
model.summary()

# =======================================
# STEP 7: Callbacks
# =======================================
os.makedirs("models", exist_ok=True)
checkpoint_path = "models/best_bilstm_v2.h5"
callbacks = [
    ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=7, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
]

# =======================================
# STEP 8: Train Model
# =======================================
history = model.fit(
    X_train_pad, y_train_cat,
    validation_data=(X_val_pad, y_val_cat),
    epochs=25,
    batch_size=64,
    callbacks=callbacks,
    class_weight=class_weights_dict,
    verbose=1
)

# =======================================
# STEP 9: Evaluate
# =======================================
if os.path.exists(checkpoint_path):
    model.load_weights(checkpoint_path)

y_pred_prob = model.predict(X_val_pad)
y_pred = np.argmax(y_pred_prob, axis=1)

print("\nClassification Report:\n")
print(classification_report(y_val, y_pred, target_names=le.classes_, zero_division=0))

cm = confusion_matrix(y_val, y_pred)
print("Confusion Matrix shape:", cm.shape)

# =======================================
# STEP 10: Plot Training Curves
# =======================================
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend(); plt.title("Loss Over Epochs")
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend(); plt.title("Accuracy Over Epochs")
plt.show()
```

```python
# ====================================================
# IMPROVED BiLSTM v3  —  Focal Loss + LR tuning + BN
# ====================================================
#Focal Loss (solves class imbalance)
#Batch Normalization (stabilizes training)
#Adam with lower LR (1e-4)
#Deeper architecture but better regularized
#More epochs (30)

Most stable training curves
import os, random, numpy as np, tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from datasets import load_dataset
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
import re, html, nltk, matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# -----------------------------
# Setup and Reproducibility
# -----------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# -----------------------------
# 1️⃣ Load dataset
# -----------------------------
ds = load_dataset("Tobi-Bueck/customer-support-tickets")
data = ds["train"] if "train" in ds else ds[list(ds.keys())[0]]

texts, labels = [], []
for row in data:
    body, queue = row.get("body"), row.get("queue")
    if body and queue:
        texts.append(body)
        labels.append(queue)
print(f"Loaded {len(texts)} samples")

# -----------------------------
# 2️⃣ Cleaning
# -----------------------------
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_ticket(text):
    if not isinstance(text, str): return ""
    text = html.unescape(text).lower()
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"\S+@\S+", " ", text)
    text = re.sub(r"\d+", " <num> ", text)
    text = re.sub(r"[^a-zA-Z0-9\s.,!?_-]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    text = text.replace("pls", "please").replace("asap", "as soon as possible")
    text = text.replace("cant", "cannot").replace("n't", " not")
    text = text.replace("u ", "you ").replace(" ur ", " your ")
    words = [lemmatizer.lemmatize(w) for w in text.split() if w not in stop_words]
    return " ".join(words)

texts = [clean_ticket(t) for t in texts]

# -----------------------------
# 3️⃣ Encode labels
# -----------------------------
le = LabelEncoder()
y = le.fit_transform(labels)
num_classes = len(le.classes_)
print(f"Classes: {num_classes}")

# -----------------------------
# 4️⃣ Tokenization & Split
# -----------------------------
X_train, X_val, y_train, y_val = train_test_split(texts, y, test_size=0.15, stratify=y, random_state=SEED)

vocab_size, max_len = 60000, 300
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_pad = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=max_len, padding="post", truncating="post")
X_val_pad   = pad_sequences(tokenizer.texts_to_sequences(X_val),   maxlen=max_len, padding="post", truncating="post")

y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes)
y_val_cat   = tf.keras.utils.to_categorical(y_val,   num_classes)

# -----------------------------
# 5️⃣ Class Weights
# -----------------------------
cw = class_weight.compute_class_weight("balanced", classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(cw))

# -----------------------------
# 6️⃣ Focal Loss
# -----------------------------
def categorical_focal_loss(gamma=2., alpha=0.25):
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow(1 - y_pred, gamma)
        return tf.reduce_mean(tf.reduce_sum(weight * cross_entropy, axis=1))
    return loss

# -----------------------------
# 7️⃣ Build Model
# -----------------------------
#Focal Loss (solves class imbalance)
#Batch Normalization (stabilizes training)
#Adam with lower LR (1e-4)
#Deeper architecture but better regularized
#More epochs (30)
def build_model(vocab_size, embedding_dim, max_len, num_classes):
    model = Sequential([
        Embedding(vocab_size, embedding_dim, input_length=max_len, trainable=True),
        Bidirectional(LSTM(256, return_sequences=True)),
        BatchNormalization(),
        Dropout(0.4),
        Bidirectional(LSTM(128)),
        Dense(256, activation="relu"),
        BatchNormalization(),
        Dropout(0.4),
        Dense(128, activation="relu"),
        Dense(num_classes, activation="softmax")
    ])
    return model

embedding_dim = 256
model = build_model(vocab_size, embedding_dim, max_len, num_classes)
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss=categorical_focal_loss(), metrics=["accuracy"])
model(tf.zeros((1, max_len)))  # build
model.summary()

# -----------------------------
# 8️⃣ Callbacks
# -----------------------------
os.makedirs("models", exist_ok=True)
ckpt = "models/best_bilstm_v3.h5"
callbacks = [
    ModelCheckpoint(ckpt, monitor="val_accuracy", save_best_only=True, verbose=1),
    EarlyStopping(monitor="val_accuracy", patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, verbose=1)
]

# -----------------------------
# 9️⃣ Train
# -----------------------------
history = model.fit(
    X_train_pad, y_train_cat,
    validation_data=(X_val_pad, y_val_cat),
    epochs=30,
    batch_size=64,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# -----------------------------
# 🔟 Evaluate
# -----------------------------
if os.path.exists(ckpt):
    model.load_weights(ckpt)

y_pred = np.argmax(model.predict(X_val_pad), axis=1)
print("\nClassification Report:")
print(classification_report(y_val, y_pred, target_names=le.classes_, zero_division=0))
print("Confusion Matrix shape:", confusion_matrix(y_val, y_pred).shape)

# -----------------------------
# 🔁 Curves
# -----------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend(); plt.title("Loss")
plt.subplot(1,2,2)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend(); plt.title("Accuracy")
plt.show()
```

```python
# ==========================================
# GEMINI INTEGRATION — CUSTOMER SUPPORT REPLY
# ==========================================
import os
import json
import requests

# ✅ Gemini REST API endpoint
GEMINI_API_URL = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent"

# ----------------------------------------------------------
# 1️⃣ Function to Generate Polite Customer Support Replies
# ----------------------------------------------------------
def generate_reply_with_gemini(ticket_body: str, predicted_queue: str, tone="polite, empathetic"):
    """
    Generates a short, empathetic and professional reply using Gemini API.
    """
    # ✅ Get your Gemini API key securely
    GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
    if GEMINI_API_KEY is None:
        raise ValueError("❌ Please set GEMINI_API_KEY environment variable before running this code.")

    # ✅ Prompt to send to Gemini
    prompt = f"""
You are a professional customer support assistant.

A new ticket has been assigned to the **{predicted_queue}** department.

Customer message:
---
{ticket_body}
---

Write a short (1–2 sentence) reply acknowledging the issue,
expressing empathy, and assuring the user the team will handle it soon.
Keep it professional and human-like. Tone: {tone}.
    """

    # ✅ Prepare API request payload
    payload = {
        "contents": [{"parts": [{"text": prompt}]}]
    }

    headers = {
        "Content-Type": "application/json",
        "x-goog-api-key": GEMINI_API_KEY
    }

    # ✅ Make the API request
    try:
        response = requests.post(GEMINI_API_URL, headers=headers, json=payload, timeout=20)
        response.raise_for_status()
        data = response.json()
        # Extract reply
        return data["candidates"][0]["content"]["parts"][0]["text"].strip()
    except Exception as e:
        return f"(Gemini reply unavailable: {e})"


# ----------------------------------------------------------
# 2️⃣ Example: Run a test sample
# ----------------------------------------------------------
import streamlit as st
import os
import google.generativeai as genai

# Load key from Streamlit Secrets
os.environ["GEMINI_API_KEY"] = st.secrets["GEMINI_API_KEY"]

# Configure Gemini API
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# Example ticket and queue prediction
sample_ticket = "My account got locked after multiple failed login attempts."
predicted_queue = "IT Support"

# Generate Gemini reply
reply = generate_reply_with_gemini(sample_ticket, predicted_queue)

print("\n🎫 Ticket Text:\n", sample_ticket)
print("\n📂 Predicted Queue:", predicted_queue)
print("\n💬 Gemini Reply:\n", reply)
```
